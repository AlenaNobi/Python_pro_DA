```python
import pandas as pd
```

```python
food_sample_100 = pd.read_csv('food_sample_100.csv')
food_other = pd.read_csv('food_other.csv')
food_nutrient =  pd.read_csv('food_nutrient.csv')
branded_food = pd.read_csv('branded_food.csv')
```

```python
food = pd.concat([food_sample_100, food_other], ignore_index=True)
# takhle mají indexi kazdá nechaný a když doplním ignore - tak už je jen index jeden společný
food
# toto odpovídá v sql funkci union (že je narveme na sebe)
```

```python
# ted bude podobná operace jako join v sql - spojíme dvě tabulky k sobě pomocí klíče - merge

food_merged = pd.merge(food, food_nutrient, on="fdc_id")
# pokud se každý sloupec jmenuje jinak (ten podle kterého to chci spojit), tak dám (left_on="fdc_id", right_on="id")
# toto je automaticky inner join, pokud chci left, right nebo outer dopíšu čárku, how="left")
food_merged
```

```python
branded_food.head()
branded_food.columns()
branded_food['brand_owner'].value_counts().head(3)
branded_food['branded_food_category'].value_counts().head(5)
```

```python
branded_food
# 135336 rows × 17 columns počet před spojením
food_merged_brands = pd.merge(branded_food, food_merged, on="fdc_id", how="right")
# food_merged_brands = pd.merge(food_merged, branded_food, right_on="fdc_id", left_on="fdc_id")
# toto je řešení z githubu (já dala jen obráceně)
food_merged_brands
# 129284 rows × 36 columns počet bez how 
# 129284 rows × 36 columns s left join
# 135336 rows × 36 columns s right join
```

```python
food_merged_brands = food_merged_brands.rename(columns={"name": "nutrient_name"})
# přejmenování sloupce
food_merged_brands

```

```python
food_merged_brands = food_merged_brands.drop(columns=["footnote", "min_year_acquired"])
# mazání sloupců
food_merged_brands
```

```python
food_merged_brands
```

```python
# kod od lektorky

food_merged_brands = pd.merge(food_merged, branded_food, right_on="fdc_id", left_on="fdc_id", how="left")
food_merged_brands = food_merged_brands.rename(columns={"name": "nutrient_name"})
food_merged_brands = food_merged_brands.drop(columns=['data_points', 'min', 'max', 'median', 'loq', 'min_year_acquired', 'derivation_id', 'footnote', 
                                                      'publication_date', 'subbrand_name','gtin_upc','ingredients', 'not_a_significant_source_of', 
                                                      'serving_size','serving_size_unit', 'household_serving_fulltext','data_source', 'package_weight',
                                                      'modified_date', 'available_date', 'market_country','discontinued_date', 'preparation_state_code', 
                                                      'trade_channel', 'short_description'])
```

```python
# dataframe.groupby("sloupecek_pro_sgrupovani")["sloupecek_pro_pocitani"].sum()
# toto je v podstatě group by a sum v sql
food_merged_brands.groupby("branded_food_category")["amount"].sum()


```

```python
food_merged_brands.groupby("branded_food_category")["amount"].mean()
```

```python
food_merged_brands_protein = food_merged_brands[food_merged_brands["nutrient_name"] == "Protein"]
food_merged_brands_protein.groupby("branded_food_category")["amount"].mean().sort_values(ascending=False).head(10)

```

```python
food_merged_brands_protein_agg = food_merged_brands_protein.groupby("branded_food_category")["amount"].mean()
food_merged_brands_protein_agg.sort_values(ascending=False).head(10)
```
