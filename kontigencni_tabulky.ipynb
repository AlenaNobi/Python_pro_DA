```python
import pandas as pd
```

```python
pivot_cholesterol = pd.read_csv("pivot_cholesterol.csv")
pivot_cholesterol
```

```python
# vytvořila jsem si seznam kategorií - nekonečno se píše pomocí float a inf
# bins = [0, 20, 100, float("inf")]
bins = [-float("inf"), 20, 100, float("inf")]
# 0 ukáže kategorii NaN - můžu dát třeba -1 a zahrne i nulu do nízké, nebo si vytvořit kategorii pro nulu
# -float("inf") - je mínus nekonečno
labels = ["Zadny", "Nizky ch", "Stredni", "Vysoky ch."]

pivot_cholesterol["cholesterol_category"] = pd.cut(pivot_cholesterol["amount"], bins=bins, labels=labels)
pivot_cholesterol.head(30)
```

```python
bins = [-float("inf"), 0, 20, 100, float("inf")]
labels = ["Zadny", "Nizky ch", "Stredni ch", "Vysoky ch."]

pivot_cholesterol["cholesterol_category"] = pd.cut(pivot_cholesterol["amount"], bins=bins, labels=labels)
pivot_cholesterol.head(30)
```

```python
pd.crosstab(pivot_cholesterol["branded_food_category"], pivot_cholesterol["cholesterol_category"], margins=True, normalize="index").style.format("{:.0%}")
# normalize ukáže vše v procentech a záleží co dám za to, podle toho to ukáže procenta
```

```python
data = pd.read_csv("hotel_bookings.csv")
data
```

```python
pd.crosstab(data["hotel"],
            data["is_canceled"],
            normalize="index"
            ).style.format("{:.0%}")
```

```python
data = data[data["hotel"] == "City Hotel"]

bins = [0, 7, 30, 180, float("inf")]
labels = ["Last-minute", "Short-term", "Medium-term", "Long-term"]
data["lead_time_category"] = pd.cut(data["lead_time"], bins=bins, labels=labels)

data_pivot = pd.crosstab(data["lead_time_category"], data["is_canceled"])
data_pivot.rename(columns={0: "Canceled", 1: "Ok"}, inplace=True)
data_pivot
```
