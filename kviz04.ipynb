```python
import pandas as pd
```

```python
data = pd.read_csv("penguins.txt")
data
```

```python
bins = [0, 180, 200, float('inf')]
labels = ['short', 'medium', 'long']

data['flipper_length_category'] = pd.cut(data['flipper_length_mm'], bins=bins, labels=labels)
data
```

```python
(data['flipper_length_category'] == 'long').sum()
```

```python
pd.crosstab(data["species"], data["island"])
```

```python
pd.crosstab(data["species"], data["island"], normalize="index")
```

```python
pd.crosstab(index=data["species"], columns=data["sex"], values=data["bill_depth_mm"], aggfunc="mean")
```
